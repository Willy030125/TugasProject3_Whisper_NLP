```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset, load_from_disk
from datasets import Dataset, DatasetDict
from dataclasses import dataclass
from typing import Any, Dict, List, Union
import torch, torchaudio
from transformers import WhisperFeatureExtractor, WhisperTokenizer, WhisperProcessor, WhisperForConditionalGeneration, Trainer, TrainingArguments
import evaluate
```

```python
processor = WhisperProcessor.from_pretrained("openai/whisper-base")
model = WhisperForConditionalGeneration.from_pretrained("openai/whisper-base")
```

```python
dataset = load_dataset("PolyAI/minds14", name="en-US")
```

```python
dataset['train']
```

```python
# Split the dataset into train, validation, and test sets
test_data = dataset["train"].train_test_split(test_size=0.1)
train_data = test_data["train"].train_test_split(test_size=0.1)
test_data = test_data["test"]
eval_data = train_data['test']
train_data = train_data['train']

dataset = DatasetDict({
    'train': train_data,
    'test': test_data,
    'validation': eval_data
})

dataset
```

```python
dataset['train'][0]
```

```python
processor.tokenizer.pad_token = processor.tokenizer.eos_token
processor.tokenizer.max_length = 224

model.config.max_length = 224
model.generation_config.max_length = 224
model.config.forced_decoder_ids = processor.tokenizer.get_decoder_prompt_ids(language="en", task="transcribe")
```

```python
def preprocess_data(batch):
    # Load audio file and resample to 16kHz
    speech_array, sampling_rate = torchaudio.load(batch["path"])
    resampler = torchaudio.transforms.Resample(orig_freq=sampling_rate, new_freq=16000)
    speech = resampler(speech_array).squeeze().numpy()

    inputs = processor(speech, sampling_rate=16000)
    batch["input_features"] = inputs.input_features[0].astype(np.float32)
    transcription = batch["transcription"]
    batch["labels"] = processor.tokenizer(text=transcription).input_ids
    return batch
```

```python
# Apply the preprocessing function to the dataset
dataset = dataset.map(preprocess_data, remove_columns=["path", "transcription"])
dataset
```

```python
# Data collator
@dataclass
class DataCollatorSpeechSeq2SeqWithPadding:
    processor: Any

    def __call__(self, features: List[Dict[str, Union[List[int], torch.Tensor]]]) -> Dict[str, torch.Tensor]:
        # Split inputs and labels
        input_features = [{"input_features": feature["input_features"]} for feature in features]
        batch = self.processor.feature_extractor.pad(input_features, return_tensors="pt")

        # Tokenize and pad labels
        label_features = [{"input_ids": feature["labels"]} for feature in features]
        labels_batch = self.processor.tokenizer.pad(label_features, return_tensors="pt")

        # Replace padding with -100 to ignore during loss calculation
        labels = labels_batch["input_ids"].masked_fill(labels_batch.attention_mask.ne(1), -100)

        batch["labels"] = labels
        return batch
```

```python
data_collator = DataCollatorSpeechSeq2SeqWithPadding(processor=processor)
```

```python
wer_metric = evaluate.load("wer")
```

```python
torch.tensor(dataset['train'][0]['input_features']).shape
```

```python
def compute_metrics(pred):
    pred_logits = pred.predictions[0]
    pred_ids = np.argmax(pred_logits, axis=-1)

    # Decode the predictions and labels
    pred_str = processor.batch_decode(pred_ids, skip_special_tokens=True)

    # Replace -100 with pad token id in labels
    label_ids = pred.label_ids
    label_ids[label_ids == -100] = processor.tokenizer.pad_token_id
    label_str = processor.batch_decode(label_ids, group_tokens=False, skip_special_tokens=True)

    # Compute WER
    wer = wer_metric.compute(predictions=pred_str, references=label_str)
    return {"WER": wer}
```

```python
training_args = TrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    gradient_accumulation_steps=1,
    evaluation_strategy="steps",
    num_train_epochs=2,
    learning_rate=2e-5,
    weight_decay=0.01,
    eval_steps=100,
    save_steps=400,
    logging_steps=100,
    logging_dir='./logs',
)
```

```python
# Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset["train"],
    eval_dataset=dataset["validation"],
    data_collator=data_collator,
    compute_metrics=compute_metrics
)
```

```python
trainer.train()
```

```python
metrics = trainer.evaluate()
metrics
```

```python
processor.save_pretrained('./whisper-base-minds14/')
trainer.save_model('./whisper-base-minds14/')
```
