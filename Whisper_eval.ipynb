```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset, load_from_disk
from datasets import Dataset, DatasetDict
import torch, torchaudio
from transformers import Wav2Vec2Processor, Wav2Vec2ForCTC, TrainingArguments, Trainer
from transformers import WhisperProcessor, WhisperForConditionalGeneration
import evaluate
```

```python
processor = WhisperProcessor.from_pretrained("Willy030125/Whisper-base-minds14")
model = WhisperForConditionalGeneration.from_pretrained("Willy030125/Whisper-base-minds14")
```

```python
processor.tokenizer.pad_token = processor.tokenizer.eos_token
processor.tokenizer.max_length = 224

model.config.max_length = 224
model.generation_config.max_length = 224
model.config.forced_decoder_ids = processor.tokenizer.get_decoder_prompt_ids(language="en", task="transcribe")
```

```python
model.to('cuda')
```

```python
wer_metric = evaluate.load("wer")
```

```python
def preprocess_and_infer(batch):
    # Load and resample audio file
    speech_array, sampling_rate = torchaudio.load(batch['audio']['path'])
    resampler = torchaudio.transforms.Resample(orig_freq=sampling_rate, new_freq=16000)
    speech = resampler(speech_array).squeeze().numpy()

    # Process audio and text
    input_feature = processor(speech, sampling_rate=16000, return_tensors="pt").input_features
    label = processor.tokenizer(batch['transcription'], padding="max_length", truncation=True, max_length=224).input_ids
    batch["input_features"] = input_feature
    batch["labels"] = label

    # Perform inference and compute loss
    input_tensor = torch.tensor(input_feature)
    label_tensor = torch.tensor(label)
    with torch.no_grad():
      #outputs = model(input_features=input_tensor, labels=label_tensor)
      predicted_ids = model.generate(input_tensor.to('cuda'))
      pred_str = processor.batch_decode(predicted_ids, skip_special_tokens=True)[0]
      #loss = outputs.loss.item()

    batch["predictions"] = pred_str
    #batch["losses"] = loss

    # Compute WER
    wer = wer_metric.compute(predictions=[pred_str], references=[batch['english_transcription']])
    batch['wer'] = wer

    return batch
```

```python
dataset = load_dataset("PolyAI/minds14", name="en-US")
```

```python
# Split the dataset into train and validation sets
#dataset = dataset["train"].train_test_split(test_size=0.1)
#dataset
```

```python
eval_result = dataset.map(preprocess_and_infer)
```
