```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset, load_from_disk
from datasets import Dataset, DatasetDict
from dataclasses import dataclass
from typing import Any, Dict, List, Union
import torch, torchaudio
from transformers import WhisperFeatureExtractor, WhisperTokenizer, WhisperProcessor, WhisperForAudioClassification, Trainer, TrainingArguments, DataCollatorWithPadding
import evaluate
```

```python
processor = WhisperProcessor.from_pretrained("openai/whisper-base")
model = WhisperForAudioClassification.from_pretrained("openai/whisper-base", num_labels=14)
```

```python
dataset = load_dataset("PolyAI/minds14", name="en-US")
```

```python
dataset['train']
```

```python
np.unique(dataset['train']['intent_class'])
```

```python
# Split the dataset into train, validation, and test sets
test_data = dataset["train"].train_test_split(test_size=0.1)
train_data = test_data["train"].train_test_split(test_size=0.1)
test_data = test_data["test"]
eval_data = train_data['test']
train_data = train_data['train']

dataset = DatasetDict({
    'train': train_data,
    'test': test_data,
    'validation': eval_data
})

dataset
```

```python
dataset['train'][0]
```

```python
processor.tokenizer.pad_token = processor.tokenizer.eos_token
processor.tokenizer.max_length = 224
```

```python
def preprocess_data(batch):
    # Load audio file and resample to 16kHz
    speech_array, sampling_rate = torchaudio.load(batch["path"])
    resampler = torchaudio.transforms.Resample(orig_freq=sampling_rate, new_freq=16000)
    speech = resampler(speech_array).squeeze().numpy()

    inputs = processor(speech, sampling_rate=16000)
    batch["input_features"] = inputs.input_features[0].astype(np.float32)
    batch["labels"] = batch['intent_class']
    return batch
```

```python
# Apply the preprocessing function to the dataset
dataset = dataset.map(preprocess_data, remove_columns=["path", "transcription", "intent_class"])
dataset
```

```python
# Data collator
@dataclass
class DataCollatorWithPadding:
    processor: Any

    def __call__(self, features: List[Dict[str, Union[List[int], torch.Tensor]]]) -> Dict[str, torch.Tensor]:
        input_features = [{"input_features": feature["input_features"]} for feature in features]
        batch = self.processor.feature_extractor.pad(input_features, return_tensors="pt")

        label_features = [{"input_ids": feature["labels"]} for feature in features]
        labels_batch = self.processor.tokenizer.pad(label_features, return_tensors="pt")

        labels = labels_batch["input_ids"].masked_fill(labels_batch.attention_mask.ne(1), -100)
        batch["labels"] = labels
        return batch
```

```python
data_collator = DataCollatorWithPadding(processor=processor)
```

```python
torch.tensor(dataset['train'][0]['input_features']).shape
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def compute_metrics(pred):
    labels = pred.label_ids
    predictions = pred.predictions.argmax(-1)

    # Compute accuracy, precision, recall, f1-score
    acc = accuracy_score(labels, predictions)
    precision = precision_score(labels, predictions, average='macro')
    recall = recall_score(labels, predictions, average='macro')
    f1 = f1_score(labels, predictions, average='macro')
    return {"Accuracy": acc, "Precision": precision, "Recall": recall, "F1 Score": f1}
```

```python
training_args = TrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=1,
    per_device_eval_batch_size=1,
    gradient_accumulation_steps=1,
    evaluation_strategy="epoch",
    save_strategy="epoch",
    num_train_epochs=5,
    learning_rate=2e-5,
    weight_decay=0.01,
    logging_dir='./logs',
    load_best_model_at_end=True,
)
```

```python
# Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=dataset["train"],
    eval_dataset=dataset["validation"],
    data_collator=data_collator,
    compute_metrics=compute_metrics
)
```

```python
trainer.train()
```

```python
metrics = trainer.evaluate()
metrics
```

```python
processor.save_pretrained('./whisper-base-minds14-ForClassification/')
trainer.save_model('./whisper-base-minds14-ForClassification/')
```
